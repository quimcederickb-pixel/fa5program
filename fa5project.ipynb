```python
#Main

def calculate_total(price, quantity):
    return price * quantity

def main():
    try:
        item_price = float(input("Enter the price of the item: "))

        item_quantity = int(input("Enter the quantity purchased: "))

        total = calculate_total(item_price, item_quantity)

        print(f"The total price for {item_quantity} item(s) is: ${total:.2f}")

    except ValueError:
        print("Invalid input. Please enter numeric values for price and quantity.")

if __name__ == "__main__":
    main()
```
